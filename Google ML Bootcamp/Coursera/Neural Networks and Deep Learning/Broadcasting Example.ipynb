```python
import numpy as np

A = np.array([[56.0, 0.0, 4.4, 68.0],
              [1.2, 104.0, 52.0, 8.0],
              [1.8, 135.0, 99.0, 0.9]])

print(A)
```

```python
cal = A.sum(axis=0) # 수직합산, 수평합산=1
print(cal) # apples, beef, eggs, potatoes의 calories
```

```python
percentage = 100*A/cal.reshape(1,4)
print(percentage)
# 3x4의 행렬을 1x4의 행렬로 나눔
```
