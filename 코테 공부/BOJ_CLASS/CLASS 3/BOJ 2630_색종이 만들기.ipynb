   NN(N=2k, k 1  7  )      .

             < 2> I, II, III, IV      N/2  N/2 .   I, II, III, IV                  .            ,           .

      < 3> < 1>     , < 4>     , < 5>     9   7    .



      N   (  )           .

```python
## 
N = int(input())
graph = [list(map(int, input().split())) for _ in range(N)]
white = 0
blue = 0

##     
def color(start_x, start_y, n):
    global white, blue
    #      
    tmp = graph[start_x][start_y]
    for i in range(start_x, start_x+n):
        for j in range(start_y, start_y+n):
            if tmp!=graph[i][j]:
                next_n = n//2 
                # 4  
                color(start_x, start_y, next_n)
                color(start_x+next_n, start_y, next_n)
                color(start_x, start_y+next_n, next_n)
                color(start_x+next_n, start_y+next_n, next_n)
                return
    #  1 blue+1, 0 white+1
    if tmp==1:
        blue += 1
    else: 
        white += 1

##   
color(0, 0, N)
print(white)
print(blue)
```
