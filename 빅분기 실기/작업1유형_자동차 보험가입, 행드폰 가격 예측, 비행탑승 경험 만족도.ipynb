## 자동차 보험 가입

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/insurance/train.csv")
df.head()
```

```python
## Vehicle_Age 값이 2년 이상인 사람들만 필터링 하고 그중에서 Annual_Premium 값이 전체 데이터의 중간값 이상인 사람들을 찾고, 그들의 Vintage값의 평균을 구하여라

tmp = df[df['Vehicle_Age']=='> 2 Years']
tmp1 = tmp[tmp['Annual_Premium']>=tmp['Annual_Premium'].median()]
result = tmp1['Vintage'].mean()
print(result)
```

```python
## vehicle_age에 따른 각 성별(gender)그룹의 Annual_Premium값의 평균을 구하여 아래 테이블과 동일하게 구현하라

df.pivot_table('Annual_Premium', 'Vehicle_Age', 'Gender', aggfunc='mean')
```

## 핸드폰 가격 예측

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/mobile/train.csv")
df.head()
```

```python
## price_range 의 각 value를 그룹핑하여 각 그룹의 n_cores 의 빈도가 가장높은 value와 그 빈도수를 구하여라

result = df.groupby(['price_range', 'n_cores']).count().sort_values(by='battery_power', ascending=False).groupby(['price_range']).head(1).reset_index()
result[['price_range', 'n_cores', 'battery_power']]
```

```python
## price_range 값이 3인 그룹에서 상관관계가 2번째로 높은 두 컬럼과 그 상관계수를 구하여라

tmp = df[df['price_range']==3]
tmp_corr = tmp.corr().unstack().to_frame('correlation').reset_index().sort_values(by='correlation', ascending=False)
tmp_corr = tmp_corr[tmp_corr['correlation'] != 1].drop_duplicates('correlation')
tmp_corr.iloc[1]
```

## 비행탑승 경험 만족도

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/airline/train.csv")
df.head()
```

```python
## Arrival Delay in Minutes 컬럼이 결측치인 데이터들 중 ‘neutral or dissatisfied’ 보다 ‘satisfied’의 수가 더 높은 Class는 어디 인가?

tmp = df[df['Arrival Delay in Minutes'].isnull()]
tmp = tmp.groupby(['Class', 'satisfaction']).count()[['id']].reset_index().pivot(index='Class', columns='satisfaction')
tmp[tmp['id', 'neutral or dissatisfied'] < tmp['id', 'satisfied']]
```
