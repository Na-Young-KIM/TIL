## 수질 음용성 여부 데이터

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/waters/train.csv")
df.head()
```

```python
## ph값은 상당히 많은 결측치를 포함한다. 결측치를 제외한 나머지 데이터들 중 사분위값 기준 하위 25%의 값들의 평균값은?

tmp = df.dropna(subset='ph')
q1 = tmp['ph'].quantile(0.25)
result = tmp[tmp['ph'] <= q1]['ph'].mean()
print(result)
```

## 의료 비용 예측 데이터

```python
import pandas as pd
train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/MedicalCost/train.csv")
train.head()
```

```python
## 흡연자와 비흡연자 각각 charges의 상위 10% 그룹의 평균의 차이는?

no_smoker, yes_smoker = train[train['smoker']=='no'], train[train['smoker']=='yes']
no_smoker = no_smoker[no_smoker['charges']>=no_smoker['charges'].quantile(0.9)]
yes_smoker = yes_smoker[yes_smoker['charges']>=yes_smoker['charges'].quantile(0.9)]
result = abs(no_smoker.charges.mean() - yes_smoker.charges.mean())
print(result)
```

## 킹카운티 주거지 가격예측문제 데이터

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/kingcountyprice//train.csv")
df.head()
```

```python
## bedrooms 의 빈도가 가장 높은 값을 가지는 데이터들의 price의 상위 10%와 하위 10%값의 차이를 구하여라

tmp = df[df['bedrooms']==df['bedrooms'].value_counts().index[0]]
result = tmp['price'].quantile(0.9) - tmp['price'].quantile(0.1)
print(result)
```

## 대학원 입학가능성 데이터

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/admission/train.csv")
df.head()
```

```python
## Serial No. 컬럼을 제외하고 ‘Chance of Admit’을 종속변수, 나머지 변수를 독립변수라 할때, 
## 랜덤포레스트를 통해 회귀 예측을 할 떄 변수중요도 값을 출력하라 (시드값에 따라 순서는 달라질수 있음)

from sklearn.ensemble import RandomForestRegressor

tmp = df.drop(df.columns[0], axis=1)
x = tmp.drop(df.columns[-1], axis=1)
y = tmp[df.columns[-1]]

ml = RandomForestRegressor()
ml.fit(x, y)

result = pd.DataFrame({'columns': x.columns, 'importance': ml.feature_importances_}).sort_values(by='importance', ascending=False)
print(result)
```

## 레드와인 퀄리티 예측 데이터

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/redwine/train.csv")
df.head()
```

```python
## quality 값이 3인 그룹과 8인 데이터그룹의 각 컬럼별 독립변수의 표준편차 값의 차이를 구할때 그값이 가장 큰 컬럼명을 구하여라

qual_3, qual_8 = df[df['quality']==3], df[df['quality']==8]
result = abs(qual_3.std() - qual_8.std()).to_frame('std 차이').sort_values(by='std 차이', ascending=False).index[0]
print(result)
```

## 약물 분류 데이터

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/drug/train.csv")
df.head()
```

```python
## 남성들의 연령대별 (10살씩 구분 0~9세 10~19세 …) Na_to_K값의 평균값을 구해서 데이터 프레임으로 표현하여라

df_man = df[df['Sex']=='M']
df_man['age_group'] = df_man['Age']//10*10
result = df_man.groupby('age_group').mean()['Na_to_K'].reset_index()
display(result)
```

## 센서데이터 동작유형 분류 데이터

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/muscle/train.csv")
df.head()
```

```python
## pose값에 따른 각 motion컬럼의 중간값의 가장 큰 차이를 보이는 motion컬럼은 어디이며 그값은?

tmp = df.groupby('pose').median().reset_index()
abs(tmp.iloc[0] - tmp.iloc[1]).to_frame('중간값 차이').sort_values(by='중간값 차이', ascending=False).iloc[0]
```

## 현대 차량 가격 분류문제 데이터

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/hyundai/train.csv")
df.head()
```

```python
## 정보(row수)가 가장 많은 상위 3차종의 price값의 각 평균값은?

top_model = list(df.groupby('model').count().sort_values(by='year', ascending=False).index[:3])
tmp = df[df['model'].apply(lambda x: x in top_model)]
result = tmp.groupby('model').mean()['price'].to_frame('price')
print(result)
```

## 넷플릭스 주식 데이터

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/nflx/NFLX.csv")
df.head()
```

```python
## 매년 5월달의 open가격의 평균값을 데이터 프레임으로 표현하라

df['Date'] = pd.to_datetime(df['Date'])
df['month'] = df['Date'].apply(lambda x: x.month)
df['year'] = df['Date'].apply(lambda x: x.year)

tmp = df[df['month']==5]
tmp.groupby('year').mean()['Open']
```

## nba 선수 능력치 데이터

```python
pd.set_option('display.max_columns',50)
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/nba/nba.csv",encoding='latin',sep=';')
df.head()
```

```python
## Tm 컬럼은 각 팀의 이름을 의미한다. TOR팀의 평균나이를 소수 4째 자리까지 구하여라

result = round(df[df['Tm']=='TOR']['Age'].mean(), 4)
print(result)
```

```python
## Pos 컬럼은 포지션을 의미한다. 전체 선수 중 최소나이대의 선수들을 필터하고 그들 중 가장 많은 포지션은 무엇인지 확인하라

tmp = df[df['Age']==df['Age'].min()]['Pos'].value_counts().to_frame('count')
result = tmp.sort_values(by='count', ascending=False).index[0]
print(result)
```

```python
## 선수들의 이름은 first_name+ 공백 + last_name으로 이루어져 있다. 가장 많은 first_name은 무엇이며 몇 회 발생하는지 확인하라

tmp = df['Player'].apply(lambda x: x.split(' ')[0]).value_counts().to_frame('count')
result = tmp.sort_values(by='count', ascending=False).iloc[0]
print(result)
```

```python
## PTS컬럼은 경기당 평균득점수 이다. 각포지션별로 경기당 평균득점수의 평균을 구하여라

result = df.groupby('Pos').mean()['PTS']
print(result)
```

```python
## G컬럼은 참여한 경기의 숫자이다. 각 팀별로 가장 높은 경기참여 수를 가진 선수들의 경기 참여 숫자의 평균을 구하여라

result = df.groupby('Tm').max()['G'].mean()
print(result)
```

```python
## Tm의 값이 MIA이며 Pos는 C또는 PF인 선수의 MP값의 평균은?

result = df[((df['Tm']=='MIA') & ((df['Pos']=='C') | (df['Pos']=='PF')))]['MP'].mean()
print(result)
```

```python
## 전체 데이터중 G의 평균값의 1.5배 이상인 데이터들만 추출했을때 3P값의 평균은?

result = df[df['G'] >= df['G'].mean()*1.5]['3P'].mean()
print(result)
```

```python
## Age의 평균 이상인 그룹과 평균 미만인 그룹간의 G값의 평균의 차이는?

upper_age = df[df['Age']>=df['Age'].mean()]
lower_age = df[df['Age']<df['Age'].mean()]

result = abs(upper_age['G'].mean() - lower_age['G'].mean())
print(result)
```

```python
## 평균나이가 가장 젊은 팀은 어디인가

tmp = df.groupby('Tm').mean()['Age'].to_frame('mean_age')
result = tmp.sort_values(by='mean_age').index[0]
print(result)
```

```python
## Pos그룹별 평균 MP값을 구하여라
df.groupby('Pos').mean()['MP']
```
